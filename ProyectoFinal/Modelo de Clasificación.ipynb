## Crear conjuntos para el entrenamiento, validación y pruebas

```python
muestras=df.shape[0]
print(muestras)
print(nombre.shape[0])
train_count=int(.8*muestras)
print(train_count)
validation_count=int(.1*muestras)
print(validation_count)
test_count=muestras - train_count - validation_count
print(test_count)

train=df[:train_count]
validation=df[train_count:train_count+validation_count]
test=df[train_count+validation_count:]

train_target=y[:train_count]
validation_target=y[train_count:train_count+validation_count]
test_target=y[train_count+validation_count:]
objetivos_test = objetivos[train_count+validation_count:]

train_nombres=nombre[:train_count]
validation_nombres=nombre[train_count:train_count+validation_count]
test_nombres=nombre[train_count+validation_count:]
print(test_target.shape)
print(test_nombres.shape)
```

## Crear los tensores con los que crearemos el modelo

```python
np.savez('data_train', inputs=train, targets=train_target)
np.savez('data_validation', inputs=validation, targets=validation_target)
np.savez('data_test', inputs=test, targets=test_target)
```

```python
npz = np.load('data_train.npz', allow_pickle=True)

train_inputs = npz['inputs'].astype(np.float)
train_targets = npz['targets'].astype(np.float)

npz = np.load('data_validation.npz', allow_pickle=True)

validation_inputs = npz['inputs'].astype(np.float)
validation_targets = npz['targets'].astype(np.float)

npz = np.load('data_test.npz', allow_pickle=True)

test_inputs = npz['inputs'].astype(np.float)
test_targets = npz['targets'].astype(np.float)
```

```python
print(train_inputs.shape)
print(train_targets.shape)

print(validation_inputs.shape)
print(validation_targets.shape)

print(test_inputs.shape)
print(test_targets.shape)
test_targets
```

```python
def build_model():
    model = tf.keras.Sequential([
    tf.keras.layers.Dense(500,activation='relu'),
    tf.keras.layers.Dense(9, activation='sigmoid')
    ])
    model.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model
model=build_model()
```

```python
batch_size=100
max_epochs=1000
early_stopping= tf.keras.callbacks.EarlyStopping(patience=100)

model.fit(train_inputs,
         train_targets,
         batch_size=batch_size,
         epochs=max_epochs,
         validation_data=(validation_inputs, validation_targets),
         verbose=2,
         callbacks=[early_stopping])
```

```python
testing=model.predict(test_inputs)
dftestpredictions = pd.DataFrame.from_records(testing)
testing
dftestpredictions
```
