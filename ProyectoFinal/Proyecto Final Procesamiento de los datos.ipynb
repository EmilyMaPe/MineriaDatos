## Procesamiento de los datos

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn import preprocessing
```

```python
path = 'appstore_games.csv'
df = pd.read_csv(path)
```

```python
nombre = df['Name']
df = df.drop(['URL','ID', 'Icon URL'],  axis = 1)
```

```python
generos = df['Genres'].str.split(', ')
generos
```

### Vamos a crear una columna para cada genero de juego

```python
mlb = MultiLabelBinarizer()
etiquetasDatos = mlb.fit_transform(df['Genres'].str.split(', '))
clases = mlb.classes_
print(clases)
etiquetas = pd.DataFrame(etiquetasDatos, columns=clases)
etiquetas.head()
```

Veamos cuantos juegos existen para cada categoria

```python
etiquetas.sum()
```

```python
categorias = list(etiquetas.columns.values)
plt.figure(figsize=(20,10))
f = sns.barplot(categorias, etiquetas.sum().values)
f.set_xticklabels(f.get_xticklabels(), rotation='25', ha="right");
```

Vemos que las etiquetas 'Games' y 'Strategy' tienen una gran afluencia, y esto lo podemos suponer por la naturaleza del dataset que es analizar juegos de estrategía. Por esto mismo vamos a eliminar estas etiquetas.

```python
etiquetas_generos = etiquetas.drop(["Strategy", "Games"], axis=1)
categorias = list(etiquetas_generos.columns.values)
plt.figure(figsize=(20,10))
f = sns.barplot(categorias, etiquetas_generos.sum().values)
f.set_xticklabels(f.get_xticklabels(), rotation='25', ha="right");
```

```python
etiquetas_generos
```

Ahora añadimos estas etiquetas a nuestro dataframe.

```python
df = pd.concat([df.reset_index(drop=True),etiquetas_generos.reset_index(drop=True)], axis=1)
print(df.shape)
df.head()
```

## Transformando las fechas
Usaremos las fechas de lanzamiento y última alctualización para saber cuatos dias tiene sin actualizarse el juego.
Los datos fueron recopilados el  3 de Agosto deñ 2019 por lo que tomaremos esa fecha como referencia.


```python
from datetime import datetime
date_format = "%d/%m/%Y"
dataset_date = datetime.strptime("3/8/2019", date_format)
curr_minus_orig = []

for i in range(len(df)):
    curr = df['Current Version Release Date'][i]
    orig = df['Original Release Date'][i]
    diff = datetime.strptime(curr, date_format) - datetime.strptime(orig, date_format)
    curr_minus_orig.append(diff.days)
    

df['Current minus Original'] = np.array(curr_minus_orig)
df['Original Release Date'] = np.array([ (dataset_date - datetime.strptime(date, date_format)).days for date in df['Original Release Date']])
df['Current Version Release Date'] =  np.array([ (dataset_date - datetime.strptime(date, date_format)).days for date in df['Current Version Release Date']])
```

```python
df.dtypes[df.dtypes == object]
```

Estas variables son las únicas que nos quedan que no son númericas, las transformaremos usando label encoding

```python
for f in df.columns:
    if df[f].dtype=='object':
        lbl = preprocessing.LabelEncoder()
        lbl.fit(list(df[f].values))
        df[f] = lbl.transform(list(df[f].values))
```
