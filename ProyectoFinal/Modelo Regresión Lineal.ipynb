```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn import preprocessing
```

```python
path = 'juegos.csv'
df = pd.read_csv(path)
```

```python
nombre = df['Name']
df = df.drop(['Name'],  axis = 1)
df
```

```python
df = df.dropna()
y = df['Average User Rating']
df = df.drop(['Average User Rating'],  axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.2, random_state=123)
```

## Regresión lineal

```python
X_train_preprocesada = preprocessing.normalize(X_train)
X_test_preprocesada = preprocessing.normalize(X_test)

lr = LinearRegression()
lr.fit(X_train_preprocesada, y_train)

y_train_predict = lr.predict(X_train_preprocesada)

rmse = (np.sqrt(mean_squared_error(y_train, y_train_predict)))

print("training set")
print('RMSE es {}\n'.format(rmse))

y_test_predict = lr.predict(X_test_preprocesada)
rmse = (np.sqrt(mean_squared_error(y_test, y_test_predict)))

print("testing set")
print('RMSE es {}'.format(rmse))
```
